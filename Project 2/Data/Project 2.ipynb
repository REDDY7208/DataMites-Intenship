```python
plt.figure(figsize=(40,40))
plotnumber=1
for i in data1:
    plt.subplot(5,5,plotnumber)
    sns.boxplot(x=data[i])
    plotnumber=plotnumber+1
plt.tight_layout()
```

```python
def find_boundaries(data,variables):
    Q1=data[variables].quantile(0.25)
    Q3=data[variables].quantile(0.75)
    IQR=Q3-Q1
    lowerboundary=Q1-1.5*IQR
    upperboundary=Q3+1.5*IQR
    return  lowerboundary,upperboundary
```

```python
lower_resting_blood_pressure,upper_resting_blood_pressure=find_boundaries(data,'resting_blood_pressure')
print('upper bondary for resting_blood_pressure',upper_resting_blood_pressure)
print('lower boundary for resting_blood_pressure',lower_resting_blood_pressure)
```

```python
data.loc[data['resting_blood_pressure']>170.0,'resting_blood_pressure']=np.median(data.resting_blood_pressure)
```

```python
lower_chest_pain_type,upper_chest_pain_type=find_boundaries(data,'chest_pain_type')
print('upper bondary for chest_pain_type',upper_chest_pain_type)
print('lower boundary for chest_pain_type',lower_chest_pain_type)
```

## LogisticRegression

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(x_sm,y_sm)
y_pred1=model1.predict(x_test)
```

## Evaluation of a Classification Model¶

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score
cm=confusion_matrix(y_test,y_pred1)
recall=recall_score(y_test,y_pred1,average='weighted')
precision=precision_score(y_test,y_pred1,average='weighted')
f1score=f1_score(y_test,y_pred1,average='weighted')
accuracy=accuracy_score(y_test,y_pred1)
print("CM:",cm)
print("Recall:",recall)
print("Precision:",precision)
print("F1 Score:",f1score)
print("Accuracy:",accuracy)
```
